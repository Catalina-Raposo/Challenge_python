### Importación de datos

#1. Análisis de facturación



## Ingreso por cada tienda


```python
PrecioTienda1 = tienda['Precio'].sum()
PrecioTienda2 = tienda2['Precio'].sum()
PrecioTienda3 = tienda3['Precio'].sum()
PrecioTienda4 = tienda4['Precio'].sum()
print(f'El ingreso total de la tienda es {PrecioTienda}')
print(f'El ingreso total de la tienda 2 es {PrecioTienda2}')
print(f'El ingreso total de la tienda 3 es {PrecioTienda3}')
print(f'El ingreso total de la tienda 4 es {PrecioTienda4}')
```

```python
TotalTiendas = PrecioTienda1 + PrecioTienda2 + PrecioTienda3 + PrecioTienda4
print(f'El total de ingresos de todas las tiendas es {TotalTiendas}')
```

```python
totales = {"Tienda":{PrecioTienda}, "Tienda2":{PrecioTienda2}, "Tienda3":{PrecioTienda3}, "Tienda 4":{PrecioTienda4}}
```

```python
cualTiendaTieneMayorIngreso = max(totales, key=totales.get)
tiendaMayorIngreso = totales[cualTiendaTieneMayorIngreso]
tiendaMayorIngresoString = "{:,}".format(int(list(totales[cualTiendaTieneMayorIngreso])[0])).replace(",",".")
print(f'La {cualTiendaTieneMayorIngreso} tiene mayor ingreso con {tiendaMayorIngresoString} pesos.')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], [PrecioTienda1, PrecioTienda2, PrecioTienda3, PrecioTienda4],color="black")
for x, y in zip(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], [PrecioTienda1, PrecioTienda2, PrecioTienda3, PrecioTienda4]):
    plt.text(x, y, str(y), ha='center', va='bottom', style = 'italic', color = 'red')

plt.xlabel('Tiendas', color= 'purple')
plt.ylabel('Ingresos', color = 'purple')
plt.title('Ingresos por Tienda', color = 'purple')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

print("Calificación Promedio Tienda 1:", calificacion_tienda)
print("Calificación Promedio Tienda 2:", calificacion_tienda2)
print("Calificación Promedio Tienda 3:", calificacion_tienda3)
print("Calificación Promedio Tienda 4:", calificacion_tienda4)
```

```python
promedios = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.figure(figsize=(10,6), facecolor='gray')
plt.plot(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], promedios, marker='o', color = 'purple')
plt.title("Calificación Promedio por Tienda", color = 'blue')
plt.ylabel("Promedio de Calificación", color = 'green')
plt.xlabel("Tienda", color = 'green')
plt.grid(visible = False,axis='y')
plt.show()
```

# 4. Productos más y menos vendidos

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas, start=1):
    producto_mas = t["Producto"].value_counts().idxmax()
    cantidad_mas = t["Producto"].value_counts().max()
    print(f"Producto más vendido en Tienda {i}: {producto_mas} (Cantidad: {cantidad_mas})")
```

```python
productos_mas = [
    tienda["Producto"].value_counts().idxmax(),
    tienda2["Producto"].value_counts().idxmax(),
    tienda3["Producto"].value_counts().idxmax(),
    tienda4["Producto"].value_counts().idxmax()
]
plt.figure(figsize=(10,6), facecolor='gray')
plt.plot(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], productos_mas, marker='o', color = 'purple')
plt.title("Producto mas vendido por tienda", color = 'blue')
plt.ylabel("Producto", color = 'green')
plt.xlabel("Tienda", color = 'green')
plt.grid(visible = False,axis='y')
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_tienda = tienda["Costo de envío"].mean()
envio_tienda2 = tienda2["Costo de envío"].mean()
envio_tienda3 = tienda3["Costo de envío"].mean()
envio_tienda4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio en Tienda 1:", envio_tienda)
print("Costo de envío promedio en Tienda 2:", envio_tienda2)
print("Costo de envío promedio en Tienda 3:", envio_tienda3)
print("Costo de envío promedio en Tienda 4:", envio_tienda4)
```

```python
envios = [envio_tienda, envio_tienda2, envio_tienda3, envio_tienda4]
labels = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(3,7))
plt.pie(envios, labels=labels, autopct='%1.1f%%', startangle=180)
plt.title("Costo de envío promedio por tienda", color = 'blue', fontweight = 'bold')
plt.gca().set_aspect('equal')
plt.show()
```
